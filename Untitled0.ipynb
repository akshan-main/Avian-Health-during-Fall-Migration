```r
library(dplyr)
library(tidyverse)
library(ggplot2)

data1 <- read.csv("/content/animals-1555860-supplementary.csv")
colnames(data1)[5]<-"agey"
data1$agey<-as.factor(data1$agey)
colnames(data1)[6]<-"fat"
data1$fat<-as.factor(data1$fat)
colnames(data1)[10]<-"NMRPDSI"
data1$NMRPDSI<-round(data1$NMRPDSI)
data1$NMRPDSI<-as.factor(data1$NMRPDSI)
```

```r
glm(fat~NMRPDSI * Diet + NMRPDSI * Residency + Age,data=data1,family="binomial")
```

```r
predicted.data <- data.frame(
  probability.of.fat=logistic$fitted.values,
  fat=data1$fat,diet=data1$Diet,NMRPDSI=data1$NMRPDSI,agey=data1$agey,Residency=data1$Residency)
 
predicted.data <- predicted.data[
  order(predicted.data$probability.of.fat, decreasing=FALSE),]

 
```

```r
ggplot(data=predicted.data,aes(x=NMRPDSI,y=probability.of.fat))+ geom_line(aes(group=diet,color=diet))+ facet_wrap(~Residency+agey)+labs(y="probability of a positive fat score")+ylim(0,1)
```
