```r
library(dplyr)
library(tidyverse)
library(ggplot2)

data1 <- read.csv("/content/animals-1555860-supplementary.csv")
colnames(data1)[5]<-"agey"

colnames(data1)[6]<-"fat"

colnames(data1)[10]<-"NMRPDSI"
```

```r
logistic<-glm(fat~NMRPDSI * Diet + NMRPDSI * Residency + agey,data=data1,family="binomial")
summary(logistic)
```

```r
predicted.data <- data.frame(
  probability.of.fat=logistic$fitted.values,
  fat=data1$fat,diet=data1$Diet,NMRPDSI=data1$NMRPDSI,agey=data1$agey,Residency=data1$Residency)
 
predicted.data <- predicted.data[
  order(predicted.data$probability.of.fat, decreasing=FALSE),]

 
```

```r
ggplot(data=predicted.data,aes(x=NMRPDSI,y=probability.of.fat))+ geom_line(aes(group=diet,color=diet))+ facet_wrap(~Residency+agey)+labs(y="probability of a positive fat score")+ylim(0,1)+xlim(-6,2)+theme_bw()
```
